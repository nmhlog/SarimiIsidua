```
from skimage import io ,color
from skimage.morphology import disk, dilation,erosion,opening,closing
from skimage.util import img_as_ubyte
import matplotlib.pyplot as plt
```

```
covid_lung =io.imread("https://raw.githubusercontent.com/nmhlog/SarimiIsidua/main/lab03/Lab%203%20-%20Images/Lab%203%20-%20Images/covid_lung.jpg")
girrafe = io.imread("https://raw.githubusercontent.com/nmhlog/SarimiIsidua/main/lab03/Lab%203%20-%20Images/Lab%203%20-%20Images/girrafe.png")
retina = io.imread("https://raw.githubusercontent.com/nmhlog/SarimiIsidua/main/lab03/Lab%203%20-%20Images/Lab%203%20-%20Images/retina.jpg")
smiley = io.imread("https://raw.githubusercontent.com/nmhlog/SarimiIsidua/main/lab03/Lab%203%20-%20Images/Lab%203%20-%20Images/smiley.png")
snowflakes = io.imread("https://raw.githubusercontent.com/nmhlog/SarimiIsidua/main/lab03/Lab%203%20-%20Images/Lab%203%20-%20Images/snowflakes.png")
```

```
im = color.rgb2gray(smiley)
se = disk(5)
dilated = dilation(im, se)
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(im, cmap='gray')
plt.axis('off')
plt.subplot(1,2,2)
plt.title('After Dilation')
plt.imshow(dilated, cmap='gray')
plt.axis('off')
plt.show()
```

```
se = disk(2)
eroded = erosion(im,se)
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(im, cmap='gray')
plt.axis('off')
plt.subplot(1,2,2)
plt.title('After Dilation')
plt.imshow(eroded, cmap='gray')
plt.axis('off')
plt.show()
```

```
import matplotlib.pyplot as plt
# Change image to gray
im = color.rgb2gray(snowflakes)
# You can change the structuring element
# Read skimage.morphology documentation for more
se = disk(10)
open = opening(im, se)
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(im, cmap='gray')
plt.axis('off')
plt.subplot(1,2,2)
plt.title('After Dilation')
plt.imshow(open, cmap='gray')
plt.axis('off')
plt.show()
```

```
# Change image to gray
im = color.rgb2gray(girrafe)
# You can change the structuring element (se)
# Read skimage.morphology documentation for more
se = disk(20)
closed = closing(im, se)
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(im, cmap='gray')
plt.axis('off')
plt.subplot(1,2,2)
plt.title('After Dilation')
plt.imshow(closed, cmap='gray')
plt.axis('off')
plt.show()
```

```
from skimage import io, color
from skimage.util import img_as_ubyte
from skimage.morphology import disk, white_tophat
import matplotlib.pyplot as plt
# Change image to RGB
im = color.rgb2gray(retina)
# You can change the structure element
# Read skimage.morphology documentation for more
se = disk(20)
tp = white_tophat(im, se)
# Show image
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(im, cmap='gray')
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Top-Hat Filter')
plt.imshow(tp, cmap='gray')
plt.axis('off')
plt.show()
```

```
from skimage import io, color
from skimage.util import img_as_ubyte
from skimage.morphology import disk, black_tophat
import matplotlib.pyplot as plt
# Change image to RGB
im = color.rgb2gray(covid_lung)
# You can change the structure element
# Read skimage.morphology documentation for more
se = disk(20)
bt = black_tophat(im, se)
# Show image
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(im, cmap='gray')
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Bottom-Hat Filter')
plt.imshow(bt, cmap='gray')
plt.axis('off')
plt.show()
```
