```python
from skimage import filters, morphology
R2 = filters.rank.mean(R,selem=morphology.square(9))
RGB = util.img_as_ubyte(np.zeros(img.shape))
RGB[:,:,0] = R2; RGB[:,:,1] = G; RGB[:,:,2] = B
plt.subplot(1,2,1); plt.imshow(img, cmap='gray')
plt.title("Original")
plt.subplot(1,2,2); plt.imshow(RGB, cmap='gray')
plt.title('img with smoothed R component');
plt.show()
```

```python
G2 = util.img_as_ubyte(filters.unsharp_mask(G,radius=5,amount=2))
RGB2 = util.img_as_ubyte(np.zeros(img.shape))
RGB2[:,:,0] = R; RGB2[:,:,1] = G2; RGB2[:,:,2] = B
plt.figure(figsize=(10,7))
plt.subplot(1,2,1); plt.imshow(img);
plt.title('original')
plt.subplot(1,2,2); plt.imshow(RGB2); plt.title('img with sharpened G component');
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
bird_image = cv2.imread("Lab 2 - Images\\Lab 2 - Images\\bird.jpg")
bird_image = cv2.cvtColor(bird_image, cv2.COLOR_BGR2RGB)
pixel_values = bird_image.reshape((-1, 3))
pixel_values = np.float32(pixel_values)
criteria = (cv2.TERM_CRITERIA_EPS +
cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
k = 4
_, labels, (centers) = cv2.kmeans(pixel_values, k, None,
criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
centers = np.uint8(centers)
labels = labels.flatten()
segmented_image = centers[labels.flatten()]
segmented_image =segmented_image.reshape(bird_image.shape)
plt.imshow(segmented_image)
plt.title("Color segementation with k=4")
plt.show()
```

```python
hsv = cv2.cvtColor(bird_image, cv2.COLOR_RGB2HSV)
low_blue = (55, 0, 0)
high_blue = (118, 255, 255)
mask = cv2.inRange(hsv, low_blue, high_blue)
result = cv2.bitwise_and(bird_image, bird_image,
mask=mask)
plt.subplot(1, 2, 1)
plt.title("mask")
plt.imshow(mask, cmap="gray")
plt.subplot(1, 2, 2)
plt.title("result")
plt.imshow(result)
plt.show()
```
